```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\bonal\OneDrive\Documents\Data\AnalyticsProject\CSV_file\Global_Mobile_Prices.csv")
df.head()
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df = df.dropna(subset=['price_usd'])
df
```

```python
plt.hist(df['price_usd'])
plt.xlabel('Price')
plt.ylabel('Count')
plt.show()
```

```python
ax = df.groupby('brand')['price_usd'].mean().sort_values().plot(kind='bar')
ax.set_xlabel('Brand')
ax.set_ylabel('Average Price usd')
plt.show()
```

```python
ax = df.groupby('brand')['rating'].mean().sort_values().plot(kind='bar')
ax.set_xlabel('Brand')
ax.set_ylabel('Average Rating')
plt.show()
```
